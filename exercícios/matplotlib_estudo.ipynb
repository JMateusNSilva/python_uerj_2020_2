**A INTERFACE ORIENTADA A OBJETO E A INTERFACE PYPLOT**

Existe essencialmente dois caminhos para se usar o Matplotlib:

*    Criando explicitamente figuras e eixos e chamando metodos sobre eles ("o estilo "OO" (orientado-objeto).
*    Ser dependente sobre pyplot criar automaticamente e gerenciar as figuras e eixos, e usar as funes pyplot para graficar.

```
# Script on OO-Style

x = np.linspace(0,2,100)

# Note que mesmo no estilo-OO, ns usamos '.pyplot.figure' para criar a figura.
fig, ax = plt.subplots()               # Cria uma figura ou eixo.
ax.plot(x, x, label = 'linear')        # Grafica alguma informao sobre os eixos...
ax.plot(x,x**2, label = 'quadratico')  # Grafica mais informes sobre os eixos...
ax.plot(x,x**3, label = 'cbico')      # e mais informaes sobre os eixos.
ax.set_xlabel('x label')               # Adiciona um rtulo ao eixo x.
ax.set_ylabel('y label')               # Adiciona um rtulo ao eixo y.
ax.set_title("Simple Plot")            # Add a title to the axes.
ax.legend()                            # Adiciona uma legenda
```

```
# Script on Pyplot-style

x = np.linspace(0,2,100)               # nmeros de pontos utilizados 100 entre os intervalos de 0 at 2.


plt.plot(x, x,    label = 'linear')
plt.plot(x, x**2, label = 'quadratic')
plt.plot(x, x**3, label = 'cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple plot")
plt.legend()
```

```
def my_plotter(ax, data1, data2, param_dict):
    """ A helper function to make a graph
    
    Parameters
    -----------
    ax : Axes
        The axes to draw to
    
    data1 : array
        The x data
    data2 : array
        The y data

    param_dict : dict
        Dictionary of kwargs to pass to ax.plot
    
    Returns
    -------
    out : list
        list of artists added
    """
    out = ax.plot(data1, data2, **param_dict)
    return out
```

```
data1, data2, data3, data4 = np.random.randn(4, 100)
fig, ax = plt.subplots(1,1)
my_plotter(ax, data1, data2, {'marker': 'x'})
```

```
fig, (ax1, ax2) = plt.subplots(1,2)
my_plotter(ax1, data1, data2, {'marker' : 'x'})
my_plotter(ax2, data3, data4, {'marker' : 'o'})
```

```
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(19680801)

# exemplo de dados
mu = 100                         # mdia da distribuio
sigma = 15                       # destribuio do desvio padro
x = mu + sigma * np.random.randn(437)  #randn - retorna uma amostra a partir de uma distribuio padro normal.

num_bins = 50

fig, ax = plt.subplots()

# Um histrograma dos dados
n, bins, patches = ax.hist(x, num_bins, density = True)

# adicionar uma linha que melhor se enquadra
y = ((1 / (np.sqrt(2 * np.pi) * sigma)) * np.exp(-0.5 * (1 / sigma * (bins -mu)) ** 2))
ax.plot(bins, y, '--')
ax.set_xlabel('Smarts')
ax.set_ylabel('Probability density')
ax.set_title(r'Histogram of IQ: $\mu=100$, $\sigma=15$')

#Tweak spacing to prevent clipping of ylabel
fig.tight_layout()
plt.show()
```
